## Text Label: 

```python
from tkinter import *
root=Tk() #create root window where all widgets go
root.geometry("700x350") #set size of the window
label1=Label(root, text="GUI in Python") # create label with words
label2=Label(root, text="Python Code")
label1.config(fg="green") # set foreground text color, bg: background 
label1.pack() #adds element/object to the frame
label2.pack()
root.mainloop() #start main loop
```

## Create Buttons:

```python
root1=Tk()
root1.geometry("400x150")
root1.title('Buttons')

def write_msg():
    print("Welcome Msg")
    print("Welcome msg")

def quit():
    root1.destroy()  

#create 1st button object with text
button1=Button(root1,
              text="Quit",
              fg="red",
              command=quit)
button1.pack(side=LEFT)

#create 2nd button object to display message 
button2=Button(root1,
               text="Welcome Msg",
               fg='green',
               command=write_msg) 
button2.pack(side=RIGHT) 

#create 3rd button object for heading
button3=Button(root1,
               text='Top',
               fg='green')
button3.pack(side=TOP)

#create 4th button object for footer
button4=Button(root1, 
               text='Bottom',
               fg='blue')
button4.pack(side=BOTTOM)


root1.mainloop()          
```

## Menus:

```python
root5=Tk()
root5.geometry("400x300")
root5.title("Menu")

#Create Menubar
menubar=Menu(root5)

#Adding File Menu & Commands:
file=Menu(menubar, tearoff=0)
menubar.add_cascade(label='File', menu=file)
file.add_command(label='New FIle', command=None)
file.add_command(label='Open...', command=None)
file.add_command(label='Save', command=None)
file.add_separator()
file.add_command(label='Exit', command=root.destroy)

#Adding Help Menu
helpInfo=Menu(menubar, tearoff=0)
menubar.add_cascade(label="Contact Help", menu=helpInfo)
helpInfo.add_command(label='Help Lessons', command=None)
helpInfo.add_separator()
helpInfo.add_command(label='About', command=None)

#display Menu
root5.config(menu=menubar)
root5.mainloop()
```

## Mouse Clicks:

```python
def button_pressed(event):
    var.set("Pressed at [" + str(event.x)+ "," + str (event.y) + "]")

def button_released(event):
    var.set("Released at [" + str(event.x)+ "," + str (event.y) + "]")    

def enter_window(event):
    var.set("Mouse is in the window")   
    
def exit_window(event):
    var.set("Mouse exited the window")

root6=Tk()
root6.title("Mouse Events")

frame=Frame(root6, width=300, height=300)
frame.pack()           

var=StringVar()
var.set("Mouse Events status will be displayed here") 
label=Label(root6, textvariable=var)
label.pack(side=BOTTOM)

frame.bind("<Button-1>", button_pressed)
#button 1 is the leftmost button, button 2 is the middle button,
#button 3 is the rightmost button
frame.bind("<ButtonRelease-1>", button_released) #Button 1 was released
frame.bind("<Enter>", enter_window) #Mouse pointer entered the widget
frame.bind("<Leave>", exit_window) #Mouse pointer exited the widget
   

root6.mainloop()    
```

## Dialog Boxes:

```python
import tkinter.messagebox

root7=Tk()
root7.title("Dialog Boxes")
root7.geometry("300x200")

def msg1():
    messagebox.showinfo('information', 'Please call for Support.')
    messagebox.showerror('error', 'There was an error!')
    messagebox.showwarning('warning', 'This is a warning')
    messagebox.askquestion('ask question', 'Do you want to continue?')
    messagebox.askokcancel('ok cancel', 'Are you sure?')

Button(root7, text='click me', command=msg1).pack()

root7.mainloop()    
    
```
